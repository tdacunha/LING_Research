```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = 'w'
plt.rcParams['figure.edgecolor'] = 'b'
plt.rcParams['font.family'] = 'serif'
# LOADING CONFUSABILITY AFTER FILE
C_file_after = open("/Users/TaraD/LING_Research/Confus_WP_full_after.txt", 'r')
Confus_after = C_file_after.read().split('\n\n')
print(np.shape(Confus_after))
print(type(Confus_after))
#print(Confus_after)
Confus_list_after = []
for i in Confus_after[:-1]:
    list1 = (i.split(' : '))
    list2 = [list1[0], float(list1[1])]
    Confus_list_after.append(list2)
Confus_array_after = (np.array(Confus_list_after))
print(Confus_array_after)
ind = np.argsort(Confus_array_after[:,1].astype(float))
x = np.take_along_axis(Confus_array_after[:,0],ind, axis=0)
y = np.take_along_axis(Confus_array_after[:,1].astype(float),ind, axis=0)

# LOADING CONFUSABILITY BEFORE FILE
C_file_before = open("/Users/TaraD/LING_Research/Confus_WP_full_before.txt", 'r')
Confus_before = C_file_before.read().split('\n\n')
print(np.shape(Confus_before))
print(type(Confus_before))
#print(Confus_after)
Confus_list_before = []
for i in Confus_before[:-1]:
    list1 = (i.split(' : '))
    list2 = [list1[0], float(list1[1])]
    Confus_list_before.append(list2)
Confus_array_before = (np.array(Confus_list_before))
print(Confus_array_before)
ind_before = np.argsort(Confus_array_before[:,1].astype(float))
x2 = np.take_along_axis(Confus_array_before[:,0],ind, axis=0)
y2 = np.take_along_axis(Confus_array_before[:,1].astype(float),ind, axis=0)
x_before = np.take_along_axis(Confus_array_before[:,0],ind_before, axis=0)
y_before = np.take_along_axis(Confus_array_before[:,1].astype(float),ind_before, axis=0)


# LOADING CONFUSABILITY SINGLE FULL FILE
C_file_single_full = open("/Users/TaraD/LING_Research/Confus_Single_Words_full.txt", 'r')
Confus_single_full = C_file_single_full.read().split('\n\n')
print(np.shape(Confus_single_full))
print(type(Confus_single_full))
#print(Confus)
Confus_list_single_full = []
for i in Confus_single_full[:-1]:
    list1 = (i.split(' : '))
    list2 = [list1[0], float(list1[1])]
    Confus_list_single_full.append(list2)
Confus_array_single_full = (np.array(Confus_list_single_full))
print(Confus_array_single_full)
print(len(Confus_array_after),len(Confus_array_single_full))
#ind = np.argsort(Confus_array_after[:,1].astype(float))
x_single = np.take_along_axis(Confus_array_single_full[:,0],ind, axis=0)
y_single = np.take_along_axis(Confus_array_single_full[:,1].astype(float),ind, axis=0)


# PLOTTING BAR GRAPHS

# WP AFTER
plt.figure(figsize = (14,5))
#plt.bar(x[::-1], np.log10(y[::-1]), .8, color = 'cadetblue')
plt.bar(x[::-1], y[::-1], .8, color = 'cadetblue')
plt.title("Word-Pair Confusability of Vowel Phonemes Full (Context After)", fontsize=17)
plt.xlabel('Vowel Phonemes', fontsize=15)
plt.ylabel('Confusability', fontsize=15)#plt.xticks
plt.show()

# WP BEFORE
plt.figure(figsize = (14,5))
#plt.bar(x[::-1], np.log10(y[::-1]), .8, color = 'cadetblue')
plt.bar(x_before[::-1], y_before[::-1], .8, color = 'yellowgreen')
plt.title("Word-Pair Confusability of Vowel Phonemes Full (Context Before)", fontsize=17)
plt.xlabel('Vowel Phonemes', fontsize=15)
plt.ylabel('Confusability', fontsize=15)#plt.xticks
plt.show()


# WP AFTER VS WP BEFORE
plt.figure(figsize = (14,5))
#plt.bar(x[::-1], np.log10(y[::-1]), .8, color = 'cadetblue')
#plt.plot(x[::-1], y[::-1], 's', markersize=10,color = 'cadetblue', label='Context After (Full)')
#plt.plot(x2[::-1], y2[::-1], 'D',markersize=10,color = 'yellowgreen', label='Context Before (Full)')
plt.bar(x[::-1], y[::-1], .8,color = 'cadetblue', label='Context After (Full)')
plt.bar(x2[::-1], y2[::-1], .5,color = 'yellowgreen', label='Context Before (Full)')
plt.title("Word-Pair Confusability Comparison:  Context After vs. Context Before", fontsize=17)
plt.xlabel('Vowel Phonemes', fontsize=15)
plt.ylabel('Confusability', fontsize=15)#plt.xticks
plt.legend()
plt.show()

# LOG WP AFTER VS SINGLE FULL
plt.figure(figsize = (14,5))
#plt.bar(x[::-1], np.log10(y[::-1]), .8, color = 'cadetblue')
plt.bar(x[::-1], np.log10(y[::-1]), .8, color = 'cadetblue', label='Word-Pairs (Context After) (Full)')
plt.bar(x2[::-1], np.log10(y2[::-1]), .6,color = 'yellowgreen', label='Word-Pairs (Context Before) (Full)')
plt.bar(x_single[::-1], np.log10(y_single[::-1]), .4, color = 'palevioletred', label='Single Words (Full)')
plt.title("Confusability Comparison: Single Words vs. Word-Pairs (Context After) vs. Word-Pairs(Context Before)", fontsize=17)
plt.xlabel('Vowel Phonemes', fontsize=15)
plt.ylabel('Log(Confusability)', fontsize=15)#plt.xticks
plt.legend()
plt.show()

# LOG WP AFTER VS SINGLE FULL
plt.figure(figsize = (14,7))
plt.plot(x[::-1], np.log10(y[::-1]), 's',ms=7, color = 'cadetblue', label='Word-Pairs (Context After) (Full)')
plt.plot(x2[::-1], np.log10(y2[::-1]), 's',ms=7,color = 'yellowgreen', label='Word-Pairs (Context Before) (Full)')
plt.plot(x_single[::-1], np.log10(y_single[::-1]), 's',ms=7, color = 'palevioletred', label='Single Words (Full)')
plt.title("Confusability Comparison: \nSingle Words vs. Word-Pairs (Context After) vs. Word-Pairs (Context Before)", fontsize=20)
plt.xlabel('Vowel Phonemes', fontsize=17)
plt.ylabel('Log(Confusability)', fontsize=17)#plt.xticks
plt.legend()
plt.show()

```

```python
# LOADING STABILIZING, DIRECTIONAL SELECTION, CONFUSABILITY 
import csv
import numpy as np

file = open('/Users/TaraD/LING_Research/stabSel.csv')
data = np.array(list(csv.reader(file)))
#print(data)
stabsel = []
for i in range(len(data)):
    line = data[i]
    if len(line)!=0:
        if i != 0:
            linelist = [line[0], float(line[1]), float(line[2]), float(line[3])]
            #print('yes')
        else:
            linelist = line
        stabsel.append(line)
stabsel = np.array(stabsel)
print(stabsel)
print(len(stabsel))

file2 = open('/Users/TaraD/LING_Research/DriveRate.csv')
data2 = np.array(list(csv.reader(file2)))
#print(data)
driverate = []
for i in range(len(data2)):
    #print(i)
    line = data2[i]
    if len(line)!=0:
        if i != 0:
            linelist = [line[0], float(line[1]), float(line[2]), float(line[3])]
            #print('yes')
        else:
            linelist = line
        driverate.append(linelist)
driverate = np.array(driverate)
print(driverate)
print(len(driverate))

dr_sorted = driverate[driverate[:,1].argsort()]
print("dr_sorted", dr_sorted)



#for i in Confus_array:
    #print(len(i),i)
print(Confus_array_after)

for i in Confus_array_after:
    if i[0] not in stabsel:
        print(i[0], type(i[1]))
    
```

```python
#PLOTTING STABILIZING SELECTION AND CONFUSABILITY ETC
stabsel_values = stabsel[1:,1:].astype(float)
#print(stabsel_values)
Confus_values = Confus_array_after[:,1].astype(float)
dr_values = driverate[1:,1:].astype(float)
#print(dr_values)
y = stabsel_values[:,0]
yl = stabsel_values[:,1]
y2 = stabsel_values[:,2]
yerr = stabsel_values[:,1:]
#print(yerr)
yerr1=yerr
yerr1[:,0] = yerr[:,0]-y[:]
yerr1[:,1] = yerr[:,1]-y[:]
yerr1 = yerr1.T
#print(np.shape(yerr1), np.shape(z))
#print(yerr1)
#print(len(stabsel_values), len(y))
#print(stabsel_values, y)
x = np.zeros((len(y)))
z = np.zeros((len(y)))
for i in range(0,len(y)):
    phoneme = stabsel[i+1,0]
    index = np.where(Confus_array_after[:,0]==phoneme)[0]
    index_dr = np.where(driverate[:,0]==phoneme)[0]
    
    #print(index)
    x[i] = Confus_values[index]
    z[i] = dr_values[index_dr-1,0]
z_median = np.median(z)
x_masked = x[z < z_median]
z_masked = z[z < z_median]
y_masked = y[z < z_median]
yerr1_masked = (yerr1.T[z < z_median]).T
#x = np.array(x)
#print(y)
#print(x)
#print(z)
#print(min(x), max(x), min(y), max(y))
#print(yerr1)
#print(np.shape(yerr1))
plt.figure(figsize=(12,5))
plt.errorbar(x,y, yerr=abs(yerr1), linestyle="None", c = 'lightsteelblue', elinewidth=.7)
plt.scatter(x,y, c = np.log10(z), cmap = 'bone', s=60)
plt.colorbar()
plt.title('Stabilizing Selection vs. Word-Pair Confusability Full (Context After) \n(Color Weighted by Log of Directional Selection)', fontsize = 15)
#print("yerr1[1]", yerr1)
plt.xlabel('Confusability', fontsize=13)
plt.ylabel('Stabilizing Selection', fontsize=13)
#plt.ylim()
plt.show()

plt.figure(figsize=(10,5))
plt.errorbar(x_masked,y_masked, yerr=abs(yerr1_masked), linestyle="None", c = 'lightsteelblue', elinewidth=.7)
plt.scatter(x_masked,y_masked, c = 'cadetblue', s=60)
slope,intercept = np.polyfit(x_masked,y_masked,1,full=True)[0]
plt.plot(x_masked, slope*x_masked + intercept, c='cadetblue', linewidth=1,label='slope: %s, intercept: %s' %(slope,intercept))
plt.legend(loc=4)
plt.title('Stabilizing Selection vs. Word-Pair Confusability Full (Context After) \n(Phonemes with less than median value of Directional Selection)', fontsize = 15)
#print("yerr1[1]", yerr1)
plt.xlabel('Confusability', fontsize=13)
plt.ylabel('Stabilizing Selection', fontsize=13)
#plt.ylim()
plt.show()

```
